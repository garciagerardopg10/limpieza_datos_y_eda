### 6. Demuestra que es mejor tener nombres de columnas sin espacios (notación corchetes y notación punto-variable).

```python
# Con notación corchetes funciona siempre
print(df['Runtime (Minutes)'].head())

# Con notación punto NO funciona si hay espacios o paréntesis
# df.Runtime (Minutes)  # esto causa SyntaxError

# En cambio, columnas sin espacios sí funcionan con punto:
print(df.Rating.head())
```

### 11. Imprime el número total de valores NA que haya en cada columna. Hazlo primero para la columna `Metascore` y después utiliza un ciclo for para hacerlo para todas las columnas

```python
# Primero para Metascore
print("NAs en Metascore:", df['Metascore'].isna().sum())

# Luego para todas las columnas con for loop
for col in df.columns:
    print(f"{col}: {df[col].isna().sum()} NAs")
```

### 20. Observa los ratings que te interesen y ahora filtra el dataframe con ese rating para ver cuáles son las películas con dicho rating

```python
# Películas con rating R
df[df['Rating'] == 'R']
```

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con essos valores

```python
frecuencia_ratings = df['Rating'].value_counts().reset_index()
frecuencia_ratings.columns = ['rating', 'frecuencia']
frecuencia_ratings
```

### 22. Ordena el nuevo dataframe por la variable `rating`

```python
frecuencia_ratings = frecuencia_ratings.sort_values('rating')
frecuencia_ratings
```

### 23. Crea una gráfica de barras con este nuevo dataframe ordenado

```python
plt.bar(frecuencia_ratings['rating'], frecuencia_ratings['frecuencia'])
plt.title('Frecuencia de películas por Rating')
plt.xlabel('Rating')
plt.ylabel('Número de películas')
plt.show()
```

### 24. Crea la matriz de correlación del dataframe de películas

```python
correlacion = df[['Rank', 'Year', 'runtime_minutes', 'Rating', 'Votes', 'revenue_millions', 'Metascore']].corr()
correlacion
```

### 25. Grafica la matriz de correlación utilizando `matshow()`

```python
fig, ax = plt.subplots()
cax = ax.matshow(correlacion, cmap='coolwarm')
fig.colorbar(cax)

columnas_numericas = ['Rank', 'Year', 'runtime_minutes', 'Rating', 'Votes', 'revenue_millions', 'Metascore']
ax.set_xticks(range(len(columnas_numericas)))
ax.set_yticks(range(len(columnas_numericas)))
ax.set_xticklabels(columnas_numericas, rotation=45, ha='left')
ax.set_yticklabels(columnas_numericas)

plt.title('Matriz de correlación', pad=20)
plt.tight_layout()
plt.show()
```
